```python
from matplotlib import pyplot as plt
```

```python
x=range(12) # 12 months a year
y = [3000, 3005, 3010, 2900, 2950, 3050, 3000, 3100, 2980, 2980, 2920, 3010]
```

Sometimes, it can be helpful to zoom in or out of the plot,especially if there is some detail we want to address. To zoom, 
we can use **plt.axis()** by feeding it a list as input. This list should contain:

+ The min x-value displayed

+ The max x_value displayed

+ The min y_value displayed

+ The max y_value displayed

```python
plt.plot(x,y)
plt.axis([0,12,2900,3100])
plt.xlabel('Time') # Add x-axis label
plt.ylabel('Dollars spent on coffee')  # Add y_aixs label
plt.title('My Last Twelve Years of Coffee Drinking')
```
